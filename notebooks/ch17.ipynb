```python
from modsim import *
```

```python
import pandas as pd
```

```python
data = pd.read_csv('glucose_insulin.csv', index_col='time')
data.head()
```

```python
data.plot(style='o')
decorate(xlabel='Time (minutes)',ylabel='Concentration (mg/dL)')
```

```python
I = interpolate(data.insulin)
```

```python
I(1), I(2), I(18)
```

```python
t_0 = data.index[0]
t_end = data.index[-1]
t_array = linrange(t_0, t_end)
I_array = I(t_array)
I_series = make_series(t_array, I_array)
```

```python
data.insulin.plot(style='o', alpha=0.5, label='insulin data')
I_series.plot(label='interpolated')
decorate(xlabel='Time (minutes)',ylabel='Concentration (Î¼U/mL)')
```

```python
# Exercise 17.1
# Pass an argument to interpolate to specify one of the other kinds of interpolation
# and plot the results.
I2 = interpolate(data.insulin, kind='cubic')
I2_array = I2(t_array)
I2_series = make_series(t_array, I2_array)
data.insulin.plot(style='o', alpha=0.5, label='insulin data')
I_series.plot(label='interpolated(linear)')
I2_series.plot(label='interpolated(cubic)')
decorate(xlabel='Time (minutes)',ylabel='Concentration (Î¼U/mL)')
```

```python
# Exercise 17.2
# Interpolate the glucose data and generate a plot, similar to the previous one, 
# that shows the data points and the interpolated curve evaluated at the time values 
# in t_array.
G = interpolate(data.glucose)
G_array = G(t_array)
G_series = make_series(t_array, G_array)
data.insulin.plot(style='o', alpha=0.5, label='insulin data')
I_series.plot(label='insulin interpolated')
data.glucose.plot(style='o', alpha=0.5, label='glucose data')
G_series.plot(label='glucose interpolated')
decorate(xlabel='Time (minutes)',ylabel='Concentration (Î¼U/dL)')
```
